# **Yapay Sinir Ağları ANN**

                Temel Yapı

        Girdi Katmanı (Girdi Katmanı) : 
Bu katman, ağın dış dünyadan kapsanan veriyi temsil eder. Örneğin, bir resim üzerinde çalışan bir ağda ona bir piksel bir girmiş olabilir.

        Gizli Katmanlar (Gizli Katmanlar) : 
Girdi verisi, bir veya daha fazla gizli katmanda işlenir. Gizli katmanlardaki her yerde, kendisine gelen girdileri ağırlıklandırır, bir aktivasyon fonksiyonunun dağılımını hesaplar. Gizli katmanlar, ağın öğrenildiği ve analiz edildiği konumun verisini verir.

        Çıktı Katmanı (Çıkış Katmanı) : 
Son olarak veriler işlendikten sonra, geniş kapsamlı bu işlenen veriyi dış dünyaya bir sonuç olarak sunar. Bu, bir sorun için bir kategori olabilir veya bir regresyon sorunu için bir sayı olabilir.

Yukarıdaki katmanların herbiri yapıtaşı olarak aşşağıdaki resimde görüleceği gibi nöronlardan oluşur her bir katman yapısı birçok nöron yapısının birleşiminden meydana gelir nöronun yapısı aşşağıda detaylıca anlatılacak


        Girdiler
Nöronun girişleri, ağdaki önceki katmanlardan gelen değerlerdir. Bu değerler, genellikle bir veri setinin özellikleridir (örneğin, bir görüntüdeki piksel değerleri veya bir tablodaki özellikler).

- Her nöronun aldığı girişler x₁, x₂, ..., xn gibi notasyonlarla ifade edilir.
- Bu girişler genellikle ağın öğrenmeye çalıştığı özelliklerdir ve nöron bu girdileri işler.

        Ağırlıklar(weights)
Her girişin, yani x₁, x₂, ..., xn'in bir ağırlığı (w₁, w₂, ..., wn) vardır. Ağırlıklar, nöronun ne kadar "önemli" bir girişe sahip olduğunu belirler. Yani, her bir girişin ağırlığı, o girişin nöronun çıkışı üzerinde ne kadar etkili olduğunu kontrol eder.

- Ağırlıklar öğrenilebilir parametrelerdir. Modelin eğitim süreci sırasında ağ, ağırlıkları ayarlayarak en doğru tahminleri yapmaya çalışır.
- Ağırlıklar, başta rastgele atanır, ancak eğitim sırasında geri yayılım (backprogentaiton) algoritmasıyla yeniden değer atanarak optimize edilir

         Bias (Sapma)
Bias, nöronun çıkışını kaydırmak için eklenen bir parametredir. Bias, ağırlıklara ve girişlere bağlı olmadan her nöronun çıkışını etkileyebilir. Genellikle, ağın öğrenmesini kolaylaştırmak ve daha esnek hale getirmek için kullanılır.

- Bias, her nöron için bir sabit değerdir ve genellikle  𝑏  ile temsil edilir.
- Nöronun girişlerinin toplamına eklenir ve öğrenme sırasında optimize edilir..



        Aktivasyon Fonksiyonu (Activation Function)
Net giriş hesaplandıktan sonra, bu değer aktivasyon fonksiyonuna gönderilir. Aktivasyon fonksiyonu, nöronun "aktif olup olmadığını" belirleyen matematiksel bir fonksiyondur ve çıkışı hesaplar. Nöronun davranışını etkileyen önemli bir bileşendir.

- Aktivasyon fonksiyonu, lineer veya non-lineer olabilir. 
- Genellikle non-lineer fonksiyonlar tercih edilir, çünkü doğrusal fonksiyonlar yalnızca lineer ilişkiler kurabilir ve karmaşık problemlerin çözülmesi için yeterli esneklik sağlamazlar.

*   --- Yaygın Aktivasyon Fonksiyonları---

1 - Sigmoid (Logistic): Çıkışları [0,1] arasında sınırlar.

2 - Tanh (Hyperbolic Tangent): Çıkışları [-1,1] arasında sınırlar.

3 - ReLU (Rectified Linear Unit): Çıkışı sıfır ile sonsuz arasında sınırlar

**** Genelde bu 3 ü çok yaygın kullanılır ***

4 - Leaky ReLU: ReLU'nun modifiye edilmiş halidir; negatif değerler için küçük bir eğim vardır.

5 - Softmax: Genellikle sınıflandırma problemlerinde kullanılır ve çıkışları [0,1] arasında toplamanın 1 olduğu bir olasılık dağılımına dönüştürür.

### Early Stopping

```python
# L1 Regularizasyonu (Lasso):
# Katsayıların mutlak değerlerinin toplamına bir ceza ekler. Bu yöntem bazı katsayıları sıfıra indirger, böylece modelde gereksiz özelliklerin etkisi azaltılır ve daha yorumlanabilir sonuçlar elde edilir.

# L2 Regularizasyonu (Ridge):
# Katsayıların karelerinin toplamını ceza terimine ekler. Bu, büyük katsayıların cezalandırılmasını sağlar ve modelin stabilitesini artırır, ancak katsayıları sıfıra indirgemekten ziyade küçültür.

# Elastic Net:
# L1 ve L2 regularizasyon tekniklerinin bir kombinasyonudur. Hem bazı katsayıları sıfıra indirgemek hem de büyük katsayıları küçültmek amacıyla kullanılır.

# Dropout:
# Eğitim sırasında, belirli oranlarda rastgele nöronların devre dışı bırakılması tekniğidir. Böylece model, nöronlar arasında aşırı bağımlılık geliştirme riskinden kaçınır ve daha sağlam bir genel performans elde edilir.

# Erken Durdurma (Early Stopping):
# Modelin doğrulama (validation) verisi üzerindeki performansı izlenerek, gelişmenin durduğu noktada eğitimin sonlandırılmasıdır. Bu yöntem, gereksiz eğitim sürelerinin modelde overfitting’e yol açmasını engeller.

```
