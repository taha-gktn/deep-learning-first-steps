
        Girdiler
Nronun girileri, adaki nceki katmanlardan gelen deerlerdir. Bu deerler, genellikle bir veri setinin zellikleridir (rnein, bir grntdeki piksel deerleri veya bir tablodaki zellikler).

- Her nronun ald giriler x, x, ..., xn gibi notasyonlarla ifade edilir.
- Bu giriler genellikle an renmeye alt zelliklerdir ve nron bu girdileri iler.

        Arlklar(weights)
Her giriin, yani x, x, ..., xn'in bir arl (w, w, ..., wn) vardr. Arlklar, nronun ne kadar "nemli" bir girie sahip olduunu belirler. Yani, her bir giriin arl, o giriin nronun k zerinde ne kadar etkili olduunu kontrol eder.

- Arlklar renilebilir parametrelerdir. Modelin eitim sreci srasnda a, arlklar ayarlayarak en doru tahminleri yapmaya alr.
- Arlklar, bata rastgele atanr, ancak eitim srasnda geri yaylm (backprogentaiton) algoritmasyla yeniden deer atanarak optimize edilir

         Bias (Sapma)
Bias, nronun kn kaydrmak iin eklenen bir parametredir. Bias, arlklara ve girilere bal olmadan her nronun kn etkileyebilir. Genellikle, an renmesini kolaylatrmak ve daha esnek hale getirmek iin kullanlr.

- Bias, her nron iin bir sabit deerdir ve genellikle    ile temsil edilir.
- Nronun girilerinin toplamna eklenir ve renme srasnda optimize edilir..



        Aktivasyon Fonksiyonu (Activation Function)
Net giri hesaplandktan sonra, bu deer aktivasyon fonksiyonuna gnderilir. Aktivasyon fonksiyonu, nronun "aktif olup olmadn" belirleyen matematiksel bir fonksiyondur ve k hesaplar. Nronun davrann etkileyen nemli bir bileendir.

- Aktivasyon fonksiyonu, lineer veya non-lineer olabilir. 
- Genellikle non-lineer fonksiyonlar tercih edilir, nk dorusal fonksiyonlar yalnzca lineer ilikiler kurabilir ve karmak problemlerin zlmesi iin yeterli esneklik salamazlar.

*   --- Yaygn Aktivasyon Fonksiyonlar---

1 - Sigmoid (Logistic): klar [0,1] arasnda snrlar.

2 - Tanh (Hyperbolic Tangent): klar [-1,1] arasnda snrlar.

3 - ReLU (Rectified Linear Unit): k sfr ile sonsuz arasnda snrlar

**** Genelde bu 3  ok yaygn kullanlr ***

4 - Leaky ReLU: ReLU'nun modifiye edilmi halidir; negatif deerler iin kk bir eim vardr.

5 - Softmax: Genellikle snflandrma problemlerinde kullanlr ve klar [0,1] arasnda toplamann 1 olduu bir olaslk dalmna dntrr.

### Early Stopping

```python
# L1 Regularizasyonu (Lasso):
# Katsaylarn mutlak deerlerinin toplamna bir ceza ekler. Bu yntem baz katsaylar sfra indirger, bylece modelde gereksiz zelliklerin etkisi azaltlr ve daha yorumlanabilir sonular elde edilir.

# L2 Regularizasyonu (Ridge):
# Katsaylarn karelerinin toplamn ceza terimine ekler. Bu, byk katsaylarn cezalandrlmasn salar ve modelin stabilitesini artrr, ancak katsaylar sfra indirgemekten ziyade kltr.

# Elastic Net:
# L1 ve L2 regularizasyon tekniklerinin bir kombinasyonudur. Hem baz katsaylar sfra indirgemek hem de byk katsaylar kltmek amacyla kullanlr.

# Dropout:
# Eitim srasnda, belirli oranlarda rastgele nronlarn devre d braklmas tekniidir. Bylece model, nronlar arasnda ar bamllk gelitirme riskinden kanr ve daha salam bir genel performans elde edilir.

# Erken Durdurma (Early Stopping):
# Modelin dorulama (validation) verisi zerindeki performans izlenerek, gelimenin durduu noktada eitimin sonlandrlmasdr. Bu yntem, gereksiz eitim srelerinin modelde overfittinge yol amasn engeller.

```
